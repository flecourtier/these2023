```python
from scar.geometry.Geometry2D import Cat,Bean
import numpy as np
import matplotlib.pyplot as plt
```

```python
form = Cat()
n_bc_points = 2000
```

```python
t = np.linspace(0,1,n_bc_points)
c_t = form.c(t)
grad_c_t = form.c_prime_rot(t,theta=form.theta_normals)
grad_c_t_norm = np.linalg.norm(grad_c_t,axis=0)
normals = grad_c_t/grad_c_t_norm

fig, ax = plt.subplots()

form.plot_curve(color="black",n_pts = n_bc_points)
ax.quiver(c_t[0,::10],c_t[1,::10],normals[0,::10],normals[1,::10],color="red")

plt.xlim(form.bord_a,form.bord_b)
plt.ylim(form.bord_a2,form.bord_b2)

plt.show()
```
