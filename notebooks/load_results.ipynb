# Load Results

## Imports

```python
from scar.problem.Case import *
from scar.utils import read_config
import matplotlib.pyplot as plt
from IPython.display import Image
import os
```

## Define case

```python
casefile = "case.json"
num_config = 1
```

```python
casefile = "../testcases/"+casefile

try:
    cas = Case(casefile)
    dir_name = "../"+cas.dir_name
    case = read_config(casefile)
    for key in case:
        print(key," : ", case[key])
except:
    print("Error reading casefile")
```

## Load Model

```python
models_dir = dir_name+"models/"

# print config
try:
    configfile = models_dir+"config_"+str(num_config)+".json"
    config = read_config(configfile)

    for key in config:
        print(key," : ", config[key])
except:
    print("Error reading configfile")
```

```python
# show training
solutions_dir = dir_name+"solutions/"
model = solutions_dir+"model_"+str(num_config)+".png"
# check if model exists
if os.path.exists(model):
    display(Image(filename=model))
else:
    print("Error loading model")
```

## Load Time-Precision

```python
deg_corr = 10

time_precision_dir = dir_name+"time_precision/"
time_precision_dir += "config"+str(num_config)+"_P"+str(deg_corr)+"/"

image = time_precision_dir+"time_precision.png"
if os.path.exists(image):
    display(Image(filename=image))
else:
    print("Error loading timeprecision")
```
